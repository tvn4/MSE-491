```python
## Model building
from sklearn.linear_model import LinearRegression

#Training features that correpsond with targets
F_val_train = F_train.iloc[:,[7,5,0]]
x_train, model = [], []
y_train = T_train

# Creation of models
for n in range (3):
    x_train.append(F_val_train.iloc[:,n].values.reshape(-1, 1))
    model.append(LinearRegression().fit(x_train[n], y_train.iloc[:,n]))
```

```python
## Visualizing linear regression results
def viz_linear(x_train, y_train, model, title, xlabel, ylabel):
    plt.scatter(x_train, y_train, color='red')
    plt.plot(x_train, model.predict(x_train), color='blue')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    return

# Model 1
plt.figure()
viz_linear(x_train[0], y_train.iloc[:,0], model[0], 'Linear Regression Model 1: CDP vs TEY (r = 0.99)', 'Compressor discharge pressure (CDP) [mbar]', 'Turbine energy yield (TEY) [MWH]')

# Model 2
plt.figure()
viz_linear(x_train[1], y_train.iloc[:,1], model[1], 'Linear Regression Model 2: TIT vs CO (r = -0.71)', 'Turbine inlet temperature (TIT) [$^o$C]', 'Carbon monoxide (CO) Emissions [mg/$m^3$]')

# Model 3
plt.figure()
viz_linear(x_train[2], y_train.iloc[:,2], model[2], 'Linear Regression Model 3: AT vs NOx (r = -0.56)', 'Ambient temperature (AT) [$^o$C]', 'Nitrous Oxide (NOx) Gas Emissions [mg/$m^3$]')
```

## 2 - Model Evaluation

```python
from math import sqrt
from sklearn.metrics import r2_score, mean_squared_error as MSE

F_val_test = F_test.iloc[:,[7,5,0]]
x_test, yp_train, yp_test, r2_train, r2_test, RMSE_train, RMSE_test = ([] for i in range(7))
y_test = T_test

## Metrics calculation
for n in range(3):
#Training Set
    yp_train.append(model[n].predict(x_train[n]).astype('int64'))
    r2_train.append(r2_score(y_train.iloc[:,n], yp_train[n]))
    RMSE_train.append(sqrt(MSE(y_train.iloc[:,n], yp_train[n])))
    print('\nModel %i (Training Set): r-squared = %f, RMSE = %f' %(n+1, r2_train[n], RMSE_train[n]))

#Test Set
    x_test.append(F_val_test.iloc[:,n].values.reshape(-1, 1))
    yp_test.append(model[n].predict(x_test[n]).astype('int64'))
    r2_test.append(r2_score(y_test.iloc[:,n], yp_test[n]))
    RMSE_test.append(sqrt(MSE(y_test.iloc[:,n], yp_test[n])))
    print('\nModel %i (Test Set): r-squared = %f, RMSE = %f' %(n+1, r2_test[n], RMSE_test[n]))
```

## 3 - Different Data Splits

```python
#Dataset initialization
from sklearn.model_selection import train_test_split

test_size = [0.5, 0.3, 0.1] #test size for each model
x_train_dd, x_test_dd, y_train_dd, y_test_dd = ([] for i in range(4))

for n in range(3):
    x4_train, x4_test, y4_train, y4_test = train_test_split(F, T, test_size=test_size[n], random_state=0)
    x_train_dd.append(x4_train.iloc[:,5].values.reshape(-1, 1))
    x_test_dd.append(x4_test.iloc[:,5].values.reshape(-1, 1))
    y_train_dd.append(y4_train.iloc[:,2])
    y_test_dd.append(y4_test.iloc[:,2])
```

```python
# Model Creation - Training/Test Set Split
model_dd = []

# Creation of models
for n in range (3):
    model_dd.append(LinearRegression().fit(x_train_dd[n], y_train_dd[n]))
    
## Model Visualization
# Model 4
plt.figure()
viz_linear(x_train_dd[0], y_train_dd[0], model_dd[0], 'Linear Regression Model 4: 50/50 Set Split', 'Turbine inlet temperature (TIT) [$^o$C]', 'Carbon monoxide (CO) Emissions [mg/$m^3$]')

# Model 5
plt.figure()
viz_linear(x_train_dd[1], y_train_dd[1], model_dd[1], 'Linear Regression Model 5: 70/30 Set Split', 'Turbine inlet temperature (TIT) [$^o$C]', 'Carbon monoxide (CO) Emissions [mg/$m^3$]')

# Model 6
plt.figure()
viz_linear(x_train_dd[2], y_train_dd[2], model_dd[2], 'Linear Regression Model 6: 90/10 Set Split', 'Turbine inlet temperature (TIT) [$^o$C]', 'Carbon monoxide (CO) Emissions [mg/$m^3$]')
```

```python
# Model Evaluation
yp_train_dd, yp_test_dd, r2_train_dd, r2_test_dd, RMSE_train_dd, RMSE_test_dd = ([] for i in range(6))

for n in range(3):
#Training Set
    yp_train_dd.append(model_dd[n].predict(x_train_dd[n]).astype('int64'))
    r2_train_dd.append(r2_score(y_train_dd[n], yp_train_dd[n]))
    RMSE_train_dd.append(sqrt(MSE(y_train_dd[n], yp_train_dd[n])))
    print('\nModel %i (Training Set): r-squared = %f, RMSE = %f' %(n+4, r2_train_dd[n], RMSE_train_dd[n]))

#Test Set
    yp_test_dd.append(model_dd[n].predict(x_test_dd[n]).astype('int64'))
    r2_test_dd.append(r2_score(y_test_dd[n], yp_test_dd[n]))
    RMSE_test_dd.append(sqrt(MSE(y_test_dd[n], yp_test_dd[n])))
    print('\nModel %i (Test Set): r-squared = %f, RMSE = %f' %(n+4, r2_test_dd[n], RMSE_test_dd[n]))
```

```python
## Barchart of RMSE Values
RMSE_dd = []

for n in range(3):
    RMSE_dd.append([('Model %i (Training)' %(n+4)), RMSE_train_dd[n]])
    RMSE_dd.append([('Model %i (Test)' %(n+4)), RMSE_test_dd[n]])

RMSE_df = pd.DataFrame (RMSE_dd, columns = ['Model # (Set)', 'RMSE'])    
    
print(RMSE_df.iloc[:,0])

plt.bar(RMSE_df.iloc[:,0],RMSE_df.iloc[:,1])
```
